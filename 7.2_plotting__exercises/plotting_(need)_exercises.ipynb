```python
#Exercise 1
from matplotlib import pyplot as plt
# Some data for plotting
x = [0, 1, 2, 3, 4, 5]
y = [0, 4, 8, 12, 16, 20]
y_err = [0.0, 0.5, 1.0, 1.5, 2.0, 2.5]
x_err = .25

plt.errorbar(x, y, 
             xerr=x_err, yerr=y_err,
             color='maroon', linestyle='dashed',
             marker='o', markerfacecolor='yellow', 
             ecolor='black',capsize=5,
             label='Y values with errors')
plt.xlabel('x-values')
plt.ylabel('y-values')
plt.title('X vs Y')
plt.grid(alpha=.25)
plt.legend(loc='lower right')
plt.show()
```

```python
#Exercise 2
import numpy as np
from matplotlib import pyplot as plt
import math
from math import e

x = np.linspace(0,50)
y = e**(-x/10)*(np.sin(x))

plt.plot(x,y,color='steelblue')
plt.title(r'y = $e^{-x/10}\sin(x)$', fontsize='20')
plt.grid(alpha=.25)
plt.xlabel('x',fontsize='15',color='darkorange')
plt.ylabel('y',fontsize='15',color='darkorange')
plt.ylim([-1, 1])
```
